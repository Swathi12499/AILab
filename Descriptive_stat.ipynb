```
from datascience import *
import numpy as np
import pandas as pd
from scipy import stats

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
```

```
data = pd.read_csv("states_all.csv", sep = ',')
data
```

```
mean = data.mean()
mean
```

```
mean_data = ["SUPPORT_SERVICES_EXPENDITURE","CAPITAL_OUTLAY_EXPENDITURE","TOTAL_EXPENDITURE"]
geometric_mean = stats.gmean(data.loc[:,mean_data])
geometric_mean
```

```
describe_data = data.describe()
describe_data
```

```
harmonic_mean = stats.hmean(data.loc[:,mean_data])
harmonic_mean
```

```
first_quartile = describe_data.iloc[4]
third_quartile = describe_data.iloc[6]
inter_quartile = third_quartile - first_quartile
inter_quartile
```

```
data.boxplot(column=["TOTAL_EXPENDITURE","STATE_REVENUE","AVG_READING_8_SCORE"])
```
